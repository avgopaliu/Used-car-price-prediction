```python
X= df1[['odometer','transmission_other','transmission_automatic','fuel_gas','type_sedan','type_pickup','fuel_diesel','type_other','type_truck']]
y = df1['price']
```

```python
#,'type_sedan','type_pickup','fuel_diesel','type_other','type_truck'
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators= 600,max_depth= 80, n_jobs=-1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred, squared=False)
```

```python
r2_score(y_test, y_pred)
```
